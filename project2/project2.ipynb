```python
import pandas as pd
test=pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/spaceship_titanic/test.csv")
train=pd.read_csv("/content/drive/MyDrive/Colab_Notebooks/spaceship_titanic/train.csv")
test_ids=test["PassengerId"]
#test_ids
```

```python
train.head(5)
```

```python
def clean(data):
  data=data.drop(["PassengerId", "Name"], axis=1)
  
  nums=["Age", "RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]
  words=["HomePlanet", "CryoSleep", "Destination", "VIP"]
  for num in nums:
      data[num].fillna(data[num].mean(), inplace=True)
  for word in words:
      data[word].fillna("Unknown", inplace=True) #Inplace才會寫進去

  return data 
```

```python
import numpy as np
train_data = clean(train)
train_data=train_data[train_data["Cabin"].notna()]
cabin=np.array(list(map(lambda x:x.split("/"), train_data["Cabin"])))
#print(cabin)
train_data["Cabin_1"]=cabin[:,0]
train_data["Cabin_2"]=cabin[:,1]
train_data["Cabin_3"]=cabin[:,2]
train_data=train_data.drop(["Cabin"], axis=1)

#freq encoding
l=train_data["Cabin_1"].value_counts()
train_data["Cabin_1"] = np.array(map(lambda x:l[x],train_data["Cabin_1"]))
train_data.head(5)
```

```python
#找眾數
#print(test_data.Cabin.value_counts())

def find_mode(l):
  gm=pd.Series(l)
  return gm.mode()[0]

test_data = clean(test)
not_empty=test_data[test_data["Cabin"].notna()]

l=list(map(lambda x:x[0], not_empty["Cabin"]))
cabin_1_mode=find_mode(l)
print(cabin_1_mode)

l=list(map(lambda x:int(x[2:-2]), not_empty["Cabin"]))
cabin_2_mode=find_mode(l)
print(cabin_2_mode)

l=list(map(lambda x:x[-1], not_empty["Cabin"]))
cabin_3_mode=find_mode(l)
print(cabin_3_mode)



#填補空缺
def f1(x):
  if type(x)==str:
    return x[0]
  elif type(x)==float:
    return cabin_1_mode

def f2(x):
  if type(x)==str:
    return int(x[2:-2])
  elif type(x)==float:
    return cabin_2_mode

def f3(x):
  if type(x)==str:
    return x[-1]
  elif type(x)==float:
    return cabin_3_mode



test_data["Cabin_1"] = np.array(map(f1,test_data["Cabin"]))
test_data["Cabin_2"] = np.array(map(f2,test_data["Cabin"]))
test_data["Cabin_3"] = np.array(map(f3,test_data["Cabin"]))
test_data=test_data.drop(["Cabin"], axis=1)

#freq encoding
l=test_data["Cabin_1"].value_counts()
#print(type(l))
test_data["Cabin_1"] = np.array(map(lambda x:l[x],test_data["Cabin_1"]))
test_data.head(5)
```

```python
X1 = train_data[["HomePlanet","CryoSleep","Destination","VIP","Cabin_3"]]
X2 = train_data[["RoomService","FoodCourt","ShoppingMall","Spa","VRDeck","Cabin_1","Cabin_2"]]
Y = train_data["Transported"]
```

```python
dummy_x = pd.get_dummies(X1)
dummy_x = dummy_x.merge(X2, how='inner', left_index=True, right_index=True)
print(dummy_x.shape)
dummy_x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
sc_x = scaler.fit_transform(dummy_x)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
import numpy as np
from tqdm import tqdm
score = np.zeros((18,50))

for i in tqdm(range(2,20,1)):
  for j in tqdm(range(150,200,1)):
    rf = RandomForestClassifier(n_estimators=j,max_depth=i)
    scores = cross_val_score(rf, sc_x, Y, cv=5)
    score[i-2,j-150]=scores.mean()

m = np.argmax(score)
i, j = divmod(m, score.shape[1])
final_depth=i+2
final_estimators=j+150
```

```python
final_rf = RandomForestClassifier(n_estimators=final_estimators, max_depth=final_depth)
final_rf.fit(sc_x, Y)
```

```python
test_data1 = test_data[["HomePlanet","CryoSleep","Destination","VIP","Cabin_3"]]
test_data2 = test_data[["RoomService","FoodCourt","ShoppingMall","Spa","VRDeck","Cabin_1","Cabin_2"]]
dummy_test_data = pd.get_dummies(test_data1)
dummy_test_data = dummy_test_data.merge(test_data2, how='inner', left_index=True, right_index=True)
sc_test_data_x = scaler.transform(dummy_test_data)
submission = final_rf.predict(sc_test_data_x)
```

```python
print(submission)
submission_data = pd.DataFrame({"PassengerId": test_ids.values, "Transported": submission })
submission_data.to_csv("submission.csv", index=False)
```
